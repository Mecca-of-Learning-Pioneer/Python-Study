```python
## 환경설정

"""
한글 폰트
이미지 사이즈
예제: sin곡선 그려보기
"""

# y = sin(x), (−2π ≤ x ≤ 2π)

import matplotlib.pyplot as plt   # package for plotting
import numpy as np

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
plt.rcParams['axes.unicode_minus'] = False
# 한글폰트 전역 설정
#plt.rcParams["font.family"] = 'NanumGothicCoding' # 'NanumGothic' # 'UnDotum' # 'NanumBarunGothic' # UnDotum
# 이미지의 크기 조정
plt.rcParams["font.size"] = 12         # 그림의 글자크기 조정 12 pt
plt.rcParams["figure.figsize"] = (8,6) # 그림의 크기를 가로 8 in, 세로 6 in

x = np.linspace(-2 * np.pi, 2 * np.pi, 50) # -2π과 2π사이를 50개의 점으로 나타냄
y = np.sin(x) 
plt.plot(x,y) # 기본 플롯

plt.show()
```

```python
## 기본 플롯

"""
# Colors
b: blue; - g: green; - c: cyan; - m: magenta; - y: yellow; - k: black; - w: white

# Symbol markers
.: point; ,: pixcel; o: circle; v: down trangle; ^: up-triangle; s: square; *: star; +: +, D: diamond

# Line style
-: solid; --: dashed; -.: dash-dot; : dotted

# Line width
lw = number, 숫자에 따라 줄 굵기가 달라짐
"""
""""""
plt.plot(x, y, "r^:", label = "legend1") # r^: 은 color = 'r', marker = '^', linestyle = ':'
plt.plot(x, 0.5 * y, color = "blue", marker = 'o', linestyle = '-', lw = 3, label = "legend2")

plt.title("sin graph", fontsize = 13) # 타이틀
plt.xlabel(r"$x$", fontsize = 12) # x축 이름
plt.ylabel(r"$sin(4 \pi x)$", fontsize = 12) # y축 이름

plt.grid(True) # 그리드 표시
plt.legend(loc = "upper left") # 레전드(범례) 표시 # loc: legend position
plt.axis([-8, 8, -1.5, 1.5]) # x축 범위, y축범위 (xmin, xmax, ymin, ymax)

plt.show()
```

```python
## Subplot

# 여러개의 그림을 행렬처럼 배치

x = np.linspace(0,1,50)
y1 = np.cos(4 * np.pi * x)
y2 = np.cos(4 * np.pi * x) * np.exp(-2 * x)

plt.subplot(2, 1, 1) # 2행 1열, idnum = 1
#plt.subplot(1, 2, 1) # 1행 2열, idnum = 1
plt.plot(x,y1,"r-*", lw = 1)
plt.grid(True)
plt.ylabel(r"$cos(4 \pi x)$")


plt.subplot(2, 1, 2)  # 2행 1열, idnum=2
#plt.subplot(1, 2, 2) # 1행 2열, idnum=2
plt.plot(x,y2,"bo",lw = 1)
plt.grid(True)
plt.xlabel('x')
plt.ylabel(r"$ e^{-2x} cos(4\pi x) $")
plt.axis([0, 2, -1.5, 1.5]) # x축 범위, y축범위 (xmin, xmax, ymin, ymax)

plt.show()
```

```python
# 상자그림

np.random.seed(19)
data = np.random.normal(0, 4, 100) # 모평균 : 0, 표준편차 : 4의 정규분포, 100개의 수를 추출
plt.boxplot(data, vert = False) # vert: 가로(False)/세로(True) 상자그림, 기본 vert는 True
plt.title("Basic Plot")

plt.show()
```

```python
data2 = np.random.normal(2, 4, 100) # 모평균 : 2, 표준편차 : 4의 정규분포, 100개의 수를 추출
type(data2)
```

```python
plt.boxplot([data, data2]) # vert: 가로(False)/세로(True) 상자그림, 기본 vert는 True
plt.title("data compare")
plt.xticks([1, 2], ["data A", "data B"])

plt.show()
```

```python
# 히스토그램

plt.hist([data, data2], label = ["data A", "data B"])
plt.legend()

plt.show()
```
